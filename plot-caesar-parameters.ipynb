```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('caesar_break_parameter_trials.csv') as c:
    results = [l for l in csv.DictReader(c, quoting=csv.QUOTE_NONNUMERIC)]
results
```

```python
for r in results:
    plt.plot(sorted([k for k in r.keys() if k != 'name']), sorted([r[k] for k in r.keys() if k != 'name']))
plt.legend([r['name'] for r in results], loc='center', bbox_to_anchor=(2, 0.5))
```

```python
for r in results:
    plt.plot(sorted([k for k in [5.0, 10.0, 20.0, 30.0]]), sorted([r[k] for k in [5.0, 10.0, 20.0, 30.0]]))
plt.legend([r['name'] for r in results], loc='center', bbox_to_anchor=(2, 0.5))
```

```python
bests = {}
for l in [k for k in results[0].keys() if k != 'name']:
    bests[l] = max(results, key=lambda r: r[l])['name']
bests
```
