```python
def cadenus_letter(n, doubled='v'):
    letter = chr(n + ord('a'))
    if letter > doubled:
        letter = chr(n + ord('a') + 1)
    return letter
```

```python
letter_positions = {letter: [(r, c) for r, row in enumerate(rows) for c, char in enumerate(row) if char == letter] for letter in 'phaseeight'}
letter_positions
```

```python
solutions = [[p] for p in letter_positions['p']]
for letter in 'has': #'haseeight':
    new_solutions = []
    for solution in solutions:
        used_columns = [p[1] for p in solution]
        for position in letter_positions[letter]:
            if position[1] not in used_columns:
                new_solutions += [solution + [position]]
    solutions = new_solutions
len(solutions)
```
