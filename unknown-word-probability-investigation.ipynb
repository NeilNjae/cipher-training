```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import log10
from cipherbreak import *
from language_models import *
```

```python
fractions = [1.0]
for wordlen in range(1, 20):
    known_words = len([w for w in keywords if len(w) == wordlen])
    possible_words = 26 ** wordlen
    fractions += [known_words / possible_words]
fractions
```

```python
plt.plot(fractions)
plt.ylabel("Probability of word")
plt.xlabel("Word length")
plt.show()
```

```python
plt.plot([log10(f) for f in fractions])
plt.plot([log10(f) for f in fractions], 'bo')
plt.ylabel("Log probability of word")
plt.xlabel("Word length")
plt.show()
```

```python
plt.plot([log10(f) for f in fractions], 'bo')
plt.plot([i for i in range(2,20)], [-(i-2) for i in range(2,20)], 'g-')
plt.ylabel("Log probability of word")
plt.xlabel("Word length")
plt.show()
```
