```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import log10
from cipherbreak import *
from language_models import *
```

# What fraction of possible words are present in the dictionary?

```python
fractions = [1.0]
for wordlen in range(1, 20):
    known_words = len([w for w in keywords if len(w) == wordlen]) # Words in the dictionary
    possible_words = 26 ** wordlen
    fractions += [known_words / possible_words]
fractions
```

# Plot the fractions

```python
plt.plot(fractions)
plt.ylabel("Probability of word")
plt.xlabel("Word length")
plt.show()
```

# Log plot

```python
plt.plot([log10(f) for f in fractions])
plt.plot([log10(f) for f in fractions], 'bo')
plt.ylabel("Log probability of word")
plt.xlabel("Word length")
plt.show()
```

# Guess some lines of best fit

```python
plt.plot([log10(f) for f in fractions], 'bo')
plt.plot([i for i in range(2,20)], [-(i-2) for i in range(2,20)], 'g-')
plt.plot([i for i in range(2,20)], [-(i-3)*1.5 for i in range(2,20)], 'r-')
plt.plot([i for i in range(2,20)], [-(i-2)*1.4 for i in range(2,20)], 'r^-')
plt.ylabel("Log probability of word")
plt.xlabel("Word length")
plt.show()
```
